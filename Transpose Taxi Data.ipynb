```python
import pandas as pd
```

```python
years = ["2020", "2019"]
months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
```

```python
list_of_dfs = []
for year in years:
    for month in months:
        df = pd.read_csv("taxi_data/by_date/taxi_data_"+year+"-"+month+".csv")
        df["year"] = int(year)
        df["month"] = int(month)
        list_of_dfs.append(df)
        
taxi_zone_lookup = pd.read_csv("taxi_data/taxi_zone_lookup.csv")
```

```python
total_df = pd.concat(list_of_dfs)
total_df
```

```python
for location_id, dataframe in total_df.groupby("location_id"):
    df = dataframe[["taxi_type", "num_trips", "total_cost", "year", "month"]]
    df = df.reset_index()
    df["location_id"] = location_id
    df["day"] = 1
    df["date"] = pd.to_datetime(df[["month", "year", "day"]])
    df["zone_name"] = taxi_zone_lookup[taxi_zone_lookup["LocationID"] == location_id]["Zone"].iloc[0]
    df.to_csv("taxi_data/by_zone/taxi_data_"+str(location_id)+".csv")
```
